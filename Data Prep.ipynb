```python
def tensorbatch2tensor(image,dtype):
    "Convert PIL style `image` array to torch style image tensor."
    a = np.asarray(image)
    if a.ndim==3 : a = np.expand_dims(a,3)
    tr_list = [i for i in range(len(a.shape) - 3)] + [-1, -3, -2]
    a = np.transpose(a, tr_list)
    return torch.from_numpy(a.astype(dtype, copy=False) )
```

```python
x= tensorbatch2tensor(inpt, np.float32)
```
